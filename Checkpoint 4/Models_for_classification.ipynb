```python
databin
```

```python
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
databin_X = databin.drop(['y3'], axis=1) 
```

```python
databin_Y = databin[['y3']]
```

```python
databin_X
```

```python
databin_Y
```

##Binary Classification

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
databin_X['gender']= label_encoder.fit_transform(databin_X['gender']) 
```

```python
databin_X['race']= label_encoder.fit_transform(databin_X['race']) 
```

```python
databin_X
```

```python
tree_model = DecisionTreeClassifier()

cv_results = cross_validate(tree_model, databin_X, databin_Y, scoring = ['accuracy'])
print(cv_results)
```

```python
print(sum(cv_results["test_accuracy"]/5))
```

```python
tree_model.fit(databin_X, databin_Y)
```

```python
kn_model = KNeighborsClassifier()
cv_results = cross_validate(kn_model, databin_X, databin_Y.values.ravel(), scoring = ['accuracy'])
print(cv_results)
```

```python
print(sum(cv_results["test_accuracy"]/5))
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(solver='liblinear')
cv_results = cross_validate(lr_model, databin_X, databin_Y.values.ravel(), scoring = ['accuracy'])
print(cv_results)
```

```python
print(sum(cv_results["test_accuracy"]/5))
```
