```python
import pandas as pd
import numpy as np
```

```python
d1 = pd.read_csv('/content/drive/MyDrive/districtpolice.csv')
d2 = pd.read_csv('/content/drive/MyDrive/allegationnums.csv')
```

```python
d1.head()
```

```python
d2
```

```python
dictmap = {1527:17, 1528:20, 1529:31, 1530:31, 1531:19, 1532:25, 1533:14, 1534:7, 1535:3, 1536:4, 1537:6, 1538:31, 1539:22, 1540:5, 1541:24, 1542:16, 1543:8, 1544:18, 1545:12, 1546:11, 1547:15, 1548:10, 1549:1, 1550:9, 1551:2}
```

```python
d2 = d2.replace({"district": dictmap})
```

```python
d2
```

```python
d6 = d1.merge(d2, on = ['district'])
```

```python
d6.head()
```

```python
d6['allegations_per_officer'] = d6['allegation_count'] / d6['police_per_district']
```

```python
d6
```

```python
d6['allegations_per_capita'] = d6['allegation_count'] / d6['district_population']
```

```python
d6
```

```python
d6['allegations_per_capita_n'] = (d6['allegations_per_capita'] - d6['allegations_per_capita'].min()) / (d6['allegations_per_capita'].max() - d6['allegations_per_capita'].min())    
```

```python
d6['allegations_per_capita_n'] = d6['allegations_per_capita_n']*100
```

```python
d6
```

```python
d6 = d6.drop(columns=['allegationspercapita'])
```
